```python
def remove_pattern(*patterns, text):
    for pattern in patterns:
        matches = pattern.finditer(text)
        temp = ''
        pos_prev = 0
        pos_cur = 0
        for match in matches:
            pos_cur = match.span(0)[0]
            temp += text[pos_prev: pos_cur]
            pos_prev = match.span(0)[1]
        temp += text[pos_prev: ]
        if temp != '':
            text = temp
    return text
```
